```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.width',2000)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
dataset = pd.read_csv('NFT_Sales.csv', delimiter=',')
dataset
```

```python
dataset.drop('Sales_USD_cumsum', inplace=True, axis=1)
dataset.drop('Number_of_Sales_cumsum', inplace=True, axis=1)
dataset.drop('Active_Market_Wallets_cumsum', inplace=True, axis=1)
dataset.drop('Primary_Sales_cumsum', inplace=True, axis=1)
dataset.drop('Secondary_Sales_cumsum', inplace=True, axis=1)
dataset.drop('AverageUSD_cum', inplace=True, axis=1)
```

```python
dataset
```

```python
dataset=dataset.drop(dataset.index[1605])
dataset=dataset.drop(dataset.index[0])
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
print(dataset['Active_Market_Wallets'].describe())
plt.figure(figsize =(10, 7))
plt.ylim(-600, 1000)
plot=dataset.boxplot(['Active_Market_Wallets'])
```

```python
q1,q2,q3=dataset.Active_Market_Wallets.quantile([0.25,0.5,0.75]).values
```

```python
q9=dataset.Active_Market_Wallets.quantile(0.90)
print(q9)
```

```python
IR = q3 -q1
print(IR)
```

```python
min = q1-1.5*IR
print(min)
```

donc le vrai min de notre data est : -228.0

```python
max = q3+1.5*IR
print(max)
```

donc le vrai max est 564.0

```python
dataset['Active_Market_Wallets'] = np.where(dataset['Active_Market_Wallets'] > q9, q2, dataset['Active_Market_Wallets'])
```

```python
print(dataset['Active_Market_Wallets'].describe())
plt.figure(figsize =(10, 7))
plt.ylim(0, 1000)
plot=dataset.boxplot(['Active_Market_Wallets'])
```

```python
dataset = dataset[~((dataset['Active_Market_Wallets'] < (q1 - 1.5 * IR)) |(dataset['Active_Market_Wallets'] > (q3 + 1.5 * IR)))
```
