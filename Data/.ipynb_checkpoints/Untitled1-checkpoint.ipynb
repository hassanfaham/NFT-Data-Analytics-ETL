```python
df = pd.read_excel (r'C:\Users\Pc\Documents\Study\MST\S2\NoSQL\nfts.xlsx',sheet_name='collection')
df
```

```python
df['items']= df['items'].replace({'K': '*1e3','M': '*1e6'}, regex=True).map(pd.eval).astype(int) #replace K to numerical value
df['Owners']= df['Owners'].replace({'K': '*1e3'}, regex=True).map(pd.eval).astype(int)

df['floor price']= df["floor price"].replace(to_replace =["---"], value = 0) #replace missing value with 0

df

```
