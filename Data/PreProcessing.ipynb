```python
def detect_outliers_iqr(data):
    outliers = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers
```

```python
num_outliers=len(detect_outliers_iqr(df_nfts['last_price']))
print(num_outliers)
print(df_nfts.shape[0])
```

```python
pourcentage_oultiers = num_outliers/df_nfts.shape[0]
print(pourcentage_oultiers*100)
```

```python
def replace_outliers(data):
    nfts_outliers = detect_outliers_iqr(data)
    print("The before : Outliers from IQR method: ", nfts_outliers)
    # Computing 10th, 90th percentiles and replacing the outliers
    # tenth_percentile = np.percentile(data, 10)
    # ninetieth_percentile = np.percentile(data, 90)
    # print(ninetieth_percentile)
    # data= np.where(data<tenth_percentile, tenth_percentile, data)
    # data = np.where(data>ninetieth_percentile, ninetieth_percentile, data)

    median = np.median(data)# Replace with median
    print('median ',median)
    for outlier in detect_outliers_iqr(data):
        data = np.where(data==outlier, median, data)
    return data
```

```python
collections = df_nfts['collection'].unique()
new_last_price = []
for col in collections:
    grp=df_nfts.loc[df_nfts['collection'] == col]
    print(grp['collection'].unique()[0])
    grp['last_price']=replace_outliers(grp['last_price'])
    nfts_outliers = detect_outliers_iqr(grp['last_price'])
    print("After replacing outliers : Outliers from IQR method: ", nfts_outliers)
    print('\n')
    new_last_price.append(grp['last_price'].tolist())
    

```
